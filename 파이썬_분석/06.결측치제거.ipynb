### 결측치 제거

```python
data = {
    '이름' : ['김자바', '이자바', '박자바', '최캐드', '정캐드']
    , '학급' : ['자바반', '자바반', '자바반', '캐드반', '캐드반']
    , '국어점수' : [90, 80, 70, 100, 90]
    , '영어점수' : [70, 80, 80, 70, 90]
    , '평균' : [80.0, 80.0, 75.0, 85.0, 90.0]
    , '취미' : ['독서', '', '', '요가', '축구']
}
```

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
# index 명이 중요할 때는 값과 이름을 주고 index=False 를 적지 않는다.
df.index = ['1번', '2번', '3번', '4번', '5번']
df.index.name = '학번'
df.to_excel('student.xlsx')
df_1 = pd.read_excel('student.xlsx', index_col='학번')
df_1
```

```python
# index를 엑셀 파일로 내보낼 때 큰 의미가 없을 때는 index=False를 적는다.
df.to_excel('student2.xlsx', index=False)
df_2 = pd.read_excel('student2.xlsx')
df_2
```

```python
# 특정 컬럼의 NaN 값을 변경하는 법
# inplace = True 를 주지 않으면 변경내용이 저장되지 않는다. (데이터 변경 안전을 위해)
df_2['취미'].fillna('없음', inplace=True)
df_2
```

```python
# NaN의 값을 제거할 때는 dropna() 함수 호출
# dropna() 함수에서 axis, how 라는 매개변수를 가진다.
# df.dropna(axis=?, how=?)
# axis='index' : NaN 데이터를 기준으로 행 삭제 (default)
# axis='columns' : NaN 데이터를 기준으로 열 삭제
# how='any' : NaN의 값이 하나라도 있으면 데이터 삭제 (default)
# how='all' : 모든 값이 NaN이면 데이터 삭제

# df_2.dropna()
# df_2.dropna(how='all')
df_2.dropna(axis='columns', how='any')
```

```python
# df_2
# 특정 컬럼의 값을 NaN로 바꾸기
# numpy ▶ 수학과 관련된 여러 기능을 제공하는 라이브러리
import numpy as np
df_2['학급'] = np.nan
df_2
```

```python
df_2.dropna(axis='columns', how='all')
```
