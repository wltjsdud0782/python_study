```python
import pandas as pd
```

```python
team_df = pd.read_csv('TeamAPI.csv')
team_df
```

```python
team_df['소음 단계'] = '보통'
team_df
```

```python
# 소음 단계별 컬럼 저장

# 좋음
team_df.loc[team_df['소음(dB)'] <= 60, '소음 단계'] = '좋음'

# 나쁨
team_df.loc[team_df['소음(dB)'] >= 80, '소음 단계'] = '나쁨'

# 보통
team_df.loc[(team_df['소음(dB)'] > 60) & (team_df['소음(dB)'] < 80), '소음 단계'] = '보통'

team_df
```

```python
# 소음 데시벨 별로 시리얼번호 구하기 (숫자만 바꾸면 가능)
where=team_df['소음(dB)'] > 60
where2 = team_df['소음(dB)'] < 80
team_df['시리얼번호'][where & where2]
```
