## Index

```python
import pandas as pd
```

```python
data = {
    '이름' : ['김자바', '이자바', '박자바', '최캐드', '정캐드']
    , '학급' : ['자바반', '자바반', '자바반', '캐드반', '캐드반']
    , '국어점수' : [90, 80, 70, 100, 90]
    , '영어점수' : [70, 80, 80, 70, 90]
    , '평균' : [80.0, 80.0, 75.0, 85.0, 90.0]
    , '취미' : ['독서', '', '', '요가', '축구']
}
```

```python
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번'])
df
```

```python
# DataFrame에서 index 값만 가져오기
df.index

# index에 이름을 부여
df.index.name = 'No'

df
```

```python
# 변경된 index를 초기화
# 아래와 같은 코드를 작성하면 index가 컬럼으로 변경되므로 좋은 코드는 아님.
df.reset_index()
```

```python
# 변경된 index를 완전히 삭제하는 법
# 매개변수로 drop만 사용하면 drop 후의 모습을 보여줄 뿐, 실제로 drop되지는 않는다.
df.reset_index(drop=True)
# inplace라는 매개변수로 저장시켜야 실제로 바뀐다. (안전성을 위함)
df.reset_index(drop=True, inplace=True)
df
```

```python
# 기존 컬럼을 index로 지정하는 법
df.set_index('이름', inplace=True)
df
```

```python
# index 기준 정렬
# ascending : 오름차순
df.sort_index(ascending=False, inplace=True)
df
```
