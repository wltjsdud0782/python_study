```python
import pandas as pd
df = pd.read_excel('student.xlsx', index_col='학번')
df
```

```python
# 데이터 교체(대체)
# 데이터 저장 시 inplace 속성 추가
df['학급'].replace({'자바반' : '수학반'})
df['학급'].replace({'자바반' : '수학반', '캐드반' : '과학반'})

my_data = {
    '자바반' : '수학반'
    , '캐드반' : '과학반'
}
df['학급'].replace(my_data)
```

```python
# 이름 컬럼값 변경
df['이름'] = 'python'
df
```

```python
# 특정 조건이 만족하는 행과 열을 찾아서 데이터를 변경하기

df.loc[df['학급'] == '캐드반', '학급'] = '과학반'
df
```

```python
# 컬럼 추가 및 해당 컬럼에 데이터 추가

# 총점 컬럼 생성 및 10 데이터 추가
df['총점'] = 10
df
```

```python
# 총점 컬럼의 값을 국어와 영어 점수의 합으로 변경

df['총점'] = df['국어점수'] + df['영어점수']
df
```

```python
# Test
# 합격 여부라는 컬럼 생성 후 총점이 160점 이상이면 'pass' 미만이면 'fail' 데이터
df['합격 여부'] = 'fail'
df.loc[df['총점'] >= 160, '합격 여부'] = 'pass'
df
```
