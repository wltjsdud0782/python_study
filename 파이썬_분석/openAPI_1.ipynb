```python
import bs4
```

```python
# response.text 를 출력하면 웹 사이트의 내용을 !!문자열!!로 출력
print(response.text)
```

```python
# 문자열 데이터에서 우리가 필요한 데이터로 쉽게 접근하기 위해서
# BeautifulSoup 형태로 데이터 변환
# 아래 코드 실행 시 오류가 발생하면 'lxml' 설치 후 진행
# ▼프린트 형태는 갖지만 문자열이 아님!!
bs_data = bs4.BeautifulSoup(response.text, 'lxml-xml')
print(bs_data)
```

```python
# find('태그명') ▶ 해당 태그 찾기
# find('태그명').text ▶ 해당 태그 안 내용 찾기

# findAll('태그명') ▶ 해당 태그 모두 출력
aa = bs_data.find('resultCode')
bb = bs_data.find('resultCode').text
print(aa)
print(bb)
```

```python
# 모든 item 태그들의 정보만 가져옴
items = bs_data.findAll('item')
print(items)
```

```python
dataTime_list = []
for item in items :
    print(item.find('dataTime').text)
    dataTime_list.append(item.find('dataTime').text)

# api 로 가져온 데이터를 모두 저장 할 dictionary
dic_data = {}
dic_data['dataTime'] = dataTime_list
print(dic_data)
```
