```python
# Python3 샘플 코드 #

import requests

url = 'http://apis.data.go.kr/3210000/SeochoIaqSvc/getSeochoIaqStnInfo'
params ={'serviceKey' : 'Pq8dvT+tIR3dsc7IfeeEl/KGNryvWt3WWuHU8Gw4VxNsT97/66aRRKa/AtNeCLpHqwzPr2hiwbS3Z5JO48djqg==', 'numOfRows' : '90', 'pageNo' : '1' }

response = requests.get(url, params=params)
print(response.content)
```

```python
# import library
import requests, xmltodict, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dict_res = xmltodict.parse(response.content)
json_string = json.dumps(dict_res['response']['body']['items'], ensure_ascii=False)

jsonObj = json.loads(json_string)
df = pd.DataFrame(jsonObj['item'])
df.columns = ['시리얼번호','스테이션명','위도','경도', '제품등록일','고객사']
df
```
