```python
import pandas as pd
df = pd.read_excel('student.xlsx', index_col='학번')
df
```

```python
# index 기준 정렬
df.sort_index() # 오름차순 정렬
df.sort_index(ascending=True) # 오름차순 정렬
df.sort_index(ascending=False, inplace=True) # 내림차순 정렬
df
```

```python
# 컬럼 기준 정렬
df.sort_values('국어점수', ascending=True)
df.sort_values('영어점수', ascending=False, inplace=True)
df
```

```python
# 평균으로 내림차순 정렬 후 이름 기준으로 오름차순 정렬
df.sort_values(['평균', '이름'], ascending=[False, True])
```
